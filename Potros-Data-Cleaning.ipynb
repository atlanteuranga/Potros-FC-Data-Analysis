```python
stats = ['Statistics-ADASL 2018-19.csv', 
         'Statistics-ADASL 2019-20.csv', 
         'Statistics-UPSL Fall 2020.csv',
         'Statistics-Liga de Naciones Spring 2021.csv',
         'Statistics-UPSL Spring 2021.csv',
         'Statistics-UPSL Fall 2021.csv',
         'Statistics-ADASL 2021-22.csv', 
         'Statistics-UPSL Regional Cup Winter 2022.csv',
         'Statistics-UPSL Spring 2022.csv',
         'Statistics-ADASL Summer 2022 (2).csv', 
         'Statistics-ADASL Summer 2022.csv', 
         'Statistics-UPSL Regional Cup Summer 2022.csv']
```

```python
stat_dfs = []
for i in range(12):
    df = pd.read_csv(f'resources/{stats[i]}')
#     print(i)
#     print(f'First name: {df.iloc[len(df)-2,0]}')
#     print(f'Name: {df.iloc[len(df)-2,2]}')
    if i == 9 or i == 10:
        df.columns = df.iloc[0]
        df = df[1:]
        df = df.drop(len(df))
        df = df.drop(len(df))
        df = df.rename(columns = {'Games Played':'MP'})
        df = df.drop(df.columns[2], axis = 1)
        df = df.dropna()
        df.to_csv(f'stats {i+1}.csv')
        stat_dfs.append(df)
        continue
    if 'Games Played' in df.columns:
        df = df.rename(columns = {'Games Played':'MP'})
    if ('Own goal' == df.iloc[len(df)-2,0]) or ('Own goal' == df.iloc[len(df)-2,2]):
        df = df.drop(len(df)-1)
        df = df.drop(len(df)-1)
    else:
        df = df.drop(len(df)-1)
    df = df.drop(df.columns[2], axis = 1)
    df = df.dropna()
    df.to_csv(f'stats {i+1}.csv')
    stat_dfs.append(df)
```

```python
stats_clean = pd.concat(stat_dfs)
stats_clean.to_csv('stats.csv', index = False)
```
